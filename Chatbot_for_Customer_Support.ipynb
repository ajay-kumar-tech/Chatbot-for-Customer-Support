```python
#Load and Preprocess Dataset
import pandas as pd
import numpy as np

# Load dataset (Assuming CSV has 'query' and 'response' columns)
df = df.dropna()  # Drop missing values

# Function to tokenize inputs and outputs
def tokenize_data(data, tokenizer, max_length=128):
    input_ids = []
    attention_masks = []

    for text in data:
        encodings = tokenizer(text, truncation=True, padding="max_length", max_length=max_length, return_tensors="tf")
        input_ids.append(encodings["input_ids"][0])
        attention_masks.append(encodings["attention_mask"][0])

    return np.array(input_ids), np.array(attention_masks)

# Tokenize queries and responses
X_input_ids, X_masks = tokenize_data(df["question"].tolist(), tokenizer)
Y_input_ids, Y_masks = tokenize_data(df["answer"].tolist(), tokenizer)
```

```python
#Create Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_input_ids, Y_input_ids, test_size=0.1, random_state=42)
```

```python
# Define optimizer, loss, and metrics
optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

# Compile model
model.compile(optimizer=optimizer, loss=loss)

# Train the model
model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=3, batch_size=8)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import nltk
import re
import string
```

```python
# Data Cleaning Function
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z?.!,¿]+", " ", text)
    return text

df["question"] = df["question"].apply(clean_text)
df["answer"] = df["answer"].apply(clean_text)
```

```python
# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df["question"].tolist() + df["answer"].tolist())

vocab_size = len(tokenizer.word_index) + 1
```

```python
# Convert text to sequences
input_sequences = tokenizer.texts_to_sequences(df["question"])
output_sequences = tokenizer.texts_to_sequences(df["answer"])
```

```python
# Padding
max_length = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_length, padding='post')
output_sequences = pad_sequences(output_sequences, maxlen=max_length, padding='post')

# Convert to NumPy array
X = np.array(input_sequences)
y = np.array(output_sequences)
```

```python
from sklearn.model_selection import train_test_split

# Assuming X = input_sequences, y = output_sequences from preprocessing step
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
```

```python
class TransformerBlock(Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim)
        ])
        self.layernorm1 = LayerNormalization(epsilon=1e-6)
        self.layernorm2 = LayerNormalization(epsilon=1e-6)
        self.dropout1 = Dropout(rate)
        self.dropout2 = Dropout(rate)

    def call(self, inputs, training=None):
        attn_output = self.att(inputs, inputs)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)

# Define the Encoder
def build_encoder(vocab_size, embed_dim, num_heads, ff_dim, max_length):
    inputs = Input(shape=(max_length,))
    embedding = Embedding(vocab_size, embed_dim, mask_zero=True)(inputs)
    transformer_block = TransformerBlock(embed_dim, num_heads, ff_dim)(embedding)
    return Model(inputs, transformer_block, name="Encoder")

# Define the Decoder
def build_decoder(vocab_size, embed_dim, num_heads, ff_dim, max_length):
    inputs = Input(shape=(max_length,))
    embedding = Embedding(vocab_size, embed_dim, mask_zero=True)(inputs)
    transformer_block = TransformerBlock(embed_dim, num_heads, ff_dim)(embedding)
    outputs = Dense(vocab_size, activation="softmax")(transformer_block)
    return Model(inputs, outputs, name="Decoder")

# Define full Transformer Seq2Seq model
def build_transformer(vocab_size, embed_dim=256, num_heads=8, ff_dim=512, max_length=50):
    encoder = build_encoder(vocab_size, embed_dim, num_heads, ff_dim, max_length)
    decoder = build_decoder(vocab_size, embed_dim, num_heads, ff_dim, max_length)

    inputs_encoder = Input(shape=(max_length,))
    inputs_decoder = Input(shape=(max_length,))

    enc_output = encoder(inputs_encoder)
    dec_output = decoder(inputs_decoder)

    model = Model(inputs=[inputs_encoder, inputs_decoder], outputs=dec_output)
    return model

# Build model
embed_dim = 256
num_heads = 8
ff_dim = 512
model = build_transformer(vocab_size, embed_dim, num_heads, ff_dim, max_length)

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
# Training the model
history = model.fit([X_train, y_train], y_train, batch_size=32, epochs=5, validation_split=0.1)
```

```python
test_loss, test_accuracy = model.evaluate([X_test, y_test], y_test)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```
